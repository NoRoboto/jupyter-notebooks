### Calculo de serie de fibonacci.

Ecuación de recurrencia:

`F(n) = a*F(n-1) + b*F(n-2) + c*F(n-3)   for n >= 3 (1)`

Dónde a, b, c son constantes.

Representaciónde la ecuación (1) de forma matricial:

[Primera Matriz] = [Segunda Matriz] * [Tercera Matriz]

| F(n)   |     =   Matriz 'C'    *  | F(n-1) |

| F(n-1) |                          | F(n-2) |

| F(n-2) |                          | F(n-3) |

n'th termino:

   `F(n) = F(n-1) + F(n-2) + F(n-3), n >= 3`
    
Caso base:

    `F(0) = 0, F(1) = 1, F(2) = 1`
    
Reemplazando las entradas con 

a = 1, b = 1 y c = 1 en la fórmula anterior:

| F(n)   |  =  [ | 1 1 1 | ] ^ (n-2)   *  | F(2) |

| F(n-1) |     [ | 1 0 0 | ]              | F(1) |

| F(n-2) |     [ | 0 1 0 | ]              | F(0) |

Fuente: https://www.geeksforgeeks.org/matrix-exponentiation/

```python
import time

def multiply(a, b):
    mul = [[0, 0], [0, 0], [0, 0]]
    for i in range(0, 2):
        for j in range(0, 2):
            for k in range(0, 2):
                mul[i][j] += a[i][k]*b[k][j]
    
    for i in range(0, 2):
        for j in range(0, 2):
            a[i][j] = mul[i][j]


def power(F, n):
    M = [[1, 1], [1, 0]]
    
    for i in range(2, n + 1):
        multiply(F, M)

def findNthTerm(n):
    F = [[1, 1], [1, 0]]
    
    if(n == 0):
        return 0
    power(F, n - 1)
    
    return F[0][0]

start_time = time.clock()

for i in range(1, 201):
    print("fibonacci " + str(i), findNthTerm(i))

print(time.clock() - start_time, "seconds")
```

```r
library(factoextra)
library(readr)
library(REdaS)
library(FactoMineR)
```

```r
data1 <- read_csv("./data1.csv")
data1
```

```r
cor(data1) # Correlación
```

```r
KMOS(data1, use = "all.obs")
```

```r
bart_spher(data1) # Chi square - degrees of freedom
```

```r
res.pca <- prcomp(data1, scale = TRUE)
summary(res.pca)
```

```r
pca2 = PCA(data1) # Componentes principales
pca2$eig
```

```r
pc <- prcomp(data1) # Componentes
pc$rotation
```

```r
fviz_eig(res.pca)
```
